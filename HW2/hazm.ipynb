```python
import docx2txt
from hazm import Normalizer, SentenceTokenizer, WordTokenizer, POSTagger, Lemmatizer
import pandas as pd
import copy
import numpy as np
```

```python
gpNum = input("enter your group number: ")
```

```python
f = open("paragraphs.docx", 'rb')
```

```python
#convert the docx to txt
txt = docx2txt.process(f)
```

```python
#seperate our group text from the entire file
l = txt.split("گروه") 
gp_txt = l[int(gpNum)]
gp_txt = gp_txt[1 : ] #remove the redundant char(group number)
gp_txt
```

```python
#preprocess
normalizer = Normalizer()
normalized = normalizer.normalize(gp_txt) #convert half space to space, remove redundant empty spaces, Create the appropriate space after the comma,...
normalized = normalizer.punctuation_spacing(normalized)
normalized  
```

```python
tokenizer = SentenceTokenizer()
wordTokenizer = WordTokenizer()
tagger = POSTagger(model= 'postagger.model')

paragraph_list = normalized.split("پاراگراف")
paragraph_list = paragraph_list[1:]

paragraoh_info = pd.DataFrame(columns = ["#Sentences" , "#Words" , "#Verbs" , "#Nouns"])
token_list = verbs_list = nouns_list =  np.array([])
all_sentence = []

for paragraph in paragraph_list:
    sentence_list = list(tokenizer.tokenize(paragraph))
    verb_count = noun_count = word_count = 0
#     all_sentence = np.append(all_sentence , sentence_list[1:])
    all_sentence.append(sentence_list[1:])
    for sentence in sentence_list[1:]:
#         all_sentence.append(sentence_list[1:])
        word_list = wordTokenizer.tokenize(sentence)
        tagged_list = tagger.tag(word_list)
        
        verbs = list(filter(lambda x: x[1] == "V", tagged_list))
        nouns = list(filter(lambda x: (x[1] == "N" or x[1] == "NE"), tagged_list))
        
        verb_count += len(verbs)
        noun_count += len(nouns)
        word_count += len(word_list)
        
        verbs_list = np.append(verbs_list , verbs)
        nouns_list = np.append(nouns_list , nouns)
        token_list = np.append(token_list , tagged_list)
        
    paragraoh_info = paragraoh_info.append({"#Sentences" : len(sentence_list[1:]),
                              "#Words" : word_count , "#Verbs" : verb_count, "#Nouns" : noun_count},ignore_index=True)

len(all_sentence)
# paragraoh_info
```

```python
print("all paragraphs" , "\n" , paragraph_list[1:])
print("all sentences" , "\n" , all_sentence)
print("all token" , "\n" , token_list)
print("all verbs" , "\n" , verbs_list)
print("all nouns" , "\n" , nouns_list)
```

```python
# Getting five random sentences
from random import randint
flat_setntences = [item for sublist in all_sentence for item in sublist]
ran = randint(0,len(flat_setntences)-1)
chosen_sentences = []
for i in range(5):
    chosen_sentences.append(flat_setntences[(ran+i*2)%len(flat_setntences)])
chosen_sentences
```

```python
# Tokenize the chosen sentences from previous step
tokenizer = WordTokenizer()
chosen_tokenized = []
for i in chosen_sentences:
    chosen_tokenized.append(tokenizer.tokenize(i))
```

```python
# Finding the lemma (bon kalamt) for each token from previous step
lemmatizer = Lemmatizer()
chosen_tagged = [tagger.tag(sent) for sent in chosen_tokenized]
chosen_lemmatized = []
for sent in chosen_tagged:
    sentence_lemmatized = {w:lemmatizer.lemmatize(w,tag) for w,tag in sent}
    chosen_lemmatized.append(sentence_lemmatized)
```
